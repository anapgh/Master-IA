### 2.1. Carga y lectura

Se van a cargar los datos del dataset para poder hacer una vista rápida de que atributos y valores tenemos. 
Nos damos cuenta de que no trae columna de indice por defecto, asi que evitamos el *'index_col=0'* en la carga del csv. Además, comprobamos que su limitación son  *','*.

```python
ruta_dataset = os.path.join("dataSet", "songs_normalize.csv")
songs_dataset = pd.read_csv(ruta_dataset, sep=',', engine="python")
display(songs_dataset)
```

Tras mostrar el dataset, explicamos que significa cada campo para entrar en contexto:

* Artista(artist): Nombre del artista.
* Canción (song): Nombre de la canción.
* Duración_ms (duration_ms): Duración de la pista en milisegundos.
* Explícito (explicit): La letra o el contenido de una canción o un vídeo musical contienen uno o varios de los criterios que podrían considerarse ofensivos o inadecuados para los niños.
* Año (year): Año de lanzamiento de la pista.
* Popularidad (popularity): Cuanto más alto sea el valor, más popular es la canción.
* Bailabilidad (danceability): La bailabilidad describe lo adecuado que es un tema para bailar basándose en una combinación de elementos musicales que incluyen el tempo, la estabilidad del ritmo, la fuerza del compás y la regularidad general. Un valor de 0,0 es el menos bailable y 1,0 el más bailable.
* Energía (energia): La energía es una medida de 0,0 a 1,0 y representa una medida perceptiva de intensidad y actividad.
* Clave (key): La tonalidad de la pista. Los números enteros se asignan a los tonos utilizando la notación estándar de Pitch Class. Por ejemplo, 0 = C, 1 = C♯/D♭, 2 = D, etc. Si no se detecta ninguna tonalidad, el valor es -1.
* Sonoridad (loudness): La sonoridad general de una pista en decibelios (dB). Los valores de sonoridad se promedian en toda la pista y son útiles para comparar la sonoridad relativa de las pistas. La sonoridad es la cualidad de un sonido que es el principal correlato psicológico de la fuerza física (amplitud). Los valores suelen oscilar entre -60 y 0 db.
* Modo (mode): El modo indica la modalidad (mayor o menor) de una pista, el tipo de escala del que se deriva su contenido melódico. La mayor se representa con 1 y la menor con 0.
* Speechiness: La locuacidad detecta la presencia de palabras habladas en una pista. Cuanto más exclusivamente hablada sea la grabación (por ejemplo, un programa de entrevistas, un audiolibro o una poesía), más se acercará a 1,0 el valor del atributo. Los valores superiores a 0,66 describen pistas que probablemente estén compuestas exclusivamente por palabras habladas. Los valores entre 0,33 y 0,66 describen pistas que pueden contener tanto música como voz, ya sea en secciones o en capas, incluyendo casos como la música rap. Los valores por debajo de 0,33 representan probablemente música y otras pistas no habladas.
* Acústica (acousticness): Una medida de confianza de 0,0 a 1,0 sobre si la pista es acústica. 1,0 representa una alta confianza en que la pista es acústica.
* Instrumentalidad(instrumentalness): Predice si una pista no contiene voces. Los sonidos "Ooh" y "aah" se consideran instrumentales en este contexto. Las pistas de rap o de palabras habladas son claramente "vocales". Cuanto más se acerque el valor de instrumentalización a 1,0, mayor será la probabilidad de que la pista no tenga contenido vocal. Los valores superiores a 0,5 representan pistas instrumentales, pero la confianza es mayor a medida que el valor se acerca a 1,0.
* Liveness: Detecta la presencia de público en la grabación. Los valores de liveness más altos representan una mayor probabilidad de que la pista se haya interpretado en directo. Un valor superior a 0,8 proporciona una fuerte probabilidad de que la pista sea en directo.
* Valence: Una medida de 0,0 a 1,0 que describe la positividad musical que transmite una pista. Las pistas con alta valencia suenan más positivas (por ejemplo, felices, alegres, eufóricas), mientras que las pistas con baja valencia suenan más negativas (por ejemplo, tristes, deprimidas, enfadadas).
* Tempo: El tempo general estimado de una pista en pulsaciones por minuto (BPM). En la terminología musical, el tempo es la velocidad o el ritmo de una pieza determinada y se deriva directamente de la duración media del compás.
* Género (genre): Género de la pista.

### 2.2. Depuración y limpieza del dataset

Lo primero de todo, es dejar el dataset listo para analizarlo, para ello se va a comprobar si tiene valores nulos.
Comprobamos que no tiene, por lo que nos ahorramos el paso de eliminarlos.

```python
# Comprobar valores nulos
songs_dataset.isnull().sum()
```

Además, nos damos cuenta de que la columna 'genre' tiene varios valores, por lo que se van a separar esos valores según el genero para que sea más sencillo.

```python
# Separamos los generos
all_genres = songs_dataset['genre'].apply(lambda x : x.split(','))

# Evitamos que se dupliquen los valores
genres = pd.unique(all_genres.sum()) 
genres = pd.unique([x.strip() for x in genres])
print(f"Géneros disponibles: \n{genres}")

# Crear tabla con columnas por género
zeros = np.zeros((len(songs_dataset), len(genres)))
genres_frame = pd.DataFrame(zeros, columns=genres)
display(genres_frame.head(3))

# Obtengo las columnas que tengo
columns_genres = genres_frame.columns # lista de generos (columnas)
# Para canción, marcar género con 1
for i, genre in enumerate(songs_dataset['genre']):
    inds = columns_genres.get_indexer(genre.split('|')) 
    genres_frame.iloc[i,inds] = 1 
    
display(genres_frame.head(5))

# Unir con dataset original
songs_split_genre = songs_dataset.join(genres_frame)
display(songs_split_genre.head(5))
```

### Artistas

Se va a hacer un estudio de los artistas más famosos y con más canciones que ha ido habiendo durante esos años. Para ello, se van a tener en cuenta las columnas de 'artist', 'song' y 'popularity'.

1. Rihana
2. Drake
3. Eminem
4. Calvin Harris

```python
# Obtenemos todos los cantantes
artist_list = pd.DataFrame(songs_dataset['artist'])
#display(artist_list)

# Contamos las apariciones de cada uno y mostramos los 5 con más canciones
display(artist_list.value_counts()[:5])
```

Calculamos la popularidad de los artistas y comprobamos que coinciden exactamente con los cantantes con más canciones publicadas.

```python
# Obtenemos los 10 más populares
artist_popu= songs_split_genre.groupby('artist')[['artist','popularity']].sum().sort_values('popularity',
                                                               ascending=False).head(10)
display(artist_popu)
```

```python
# Creamos una gráfica para plasmar mejor la información
plt.figure(figsize=(12,6))
ax = sns.barplot(data = artist_popu, x = 'popularity', y = artist_popu.index ,palette='deep')
plt.ylabel('Artista', size = 15)
plt.xlabel('Puntuación de popularidad', size = 15)
plt.title('Artistas más populares', size = 20)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10, rotation = 45)
ax.bar_label(ax.containers[0],size = 10)
plt.show()
```

Para relacionar completamente las columnas utilizadas, se va a mostrar una regresión lineal par comprobar su relación.

```python
# Preparamos una tabla con las tres columnas interesadas
popu_artist = songs_dataset.groupby("artist").popularity.mean().reset_index()
songs_artist = songs_dataset.groupby("artist").size().reset_index()
artists_total = popu_artist.merge(songs_artist, on="artist")
artists_total.columns = ["artist", "popularity", "songs"]
display(artists_total.head())

# Regresión de la popularidad de las canciones en artistas
plt.figure(figsize=(10,5))
ax = sns.regplot(y="popularity", x="songs", data=artists_total)
ax.set_title("Popularidad de los artistas por el número de canciones", fontsize=15)
ax.set_xlabel("Número de canciones")
ax.set_xticks(np.arange(1, 26, 1))
ax.set_ylabel("Media de popularidad")
ax.set_yticks(np.arange(0, 91, 5))
```

Comprobando así, que de media, los artistas con más popularidad en esta lista de *Top Hits* podían sacar 25 canciones y ser muy populares, o sacar 1 y serlo también.

## 4. Relación entre variables

La correlación, es la relación matemática entre dos variables (-1 negativamente relacionadas, 1 positivamente relacionadas, 0 sin relación). 

De esta forma, vamos a comprobar si algunas de las columnas más técnicas como la instrumentación, la energia, etc, puede aportarnos más información de la obtenida anteriormente.

Lo primero de todo, vamos a hacer una correlación total de todos los valores númericos y poder comprobar cuales guardan más relación.

```python
# Calculamos la correlación entre los valores que nos interesan
corr_data = songs_dataset.drop(['year'],axis=1).corr()
display(corr_data)
```

```python
# Representación gráfica
corr_data.style.background_gradient(cmap=plt.get_cmap('RdYlGn'), axis=1)
```

Con el resultado anterior, se puede comprobar que en general no hay muchas características que guarden relación. Algunas que podrías destacarse serían las siguientes:

* loudness - energy : 0.651016
* speechiness - explicit : 0.417343	
* valence - danceability : 0.403178
